````python
#get Jenny's solution
def getprimes(rnge):
    primes = []
    n = 2
    while n < rnge + 1:
        isPrime = True
        for i in range(2, n):           #divides n by every number from 2 to itself
            if n % i == 0 and n != i:   #if both are met it is not prime,
                #
                isPrime = False
                break                   #breaks out of for loop, 
        if isPrime:                     #if prime, is added to primes lst
            primes.append(n)
        n += 1                          #goes to next n
    return primes

getprimes(1000)
````
